```python
def matrix_factorization(R, K, steps=200, learning_rate=0.01, r_lambda = 0.02):
    num_reviewers, num_beers = R.shape
    # P와 Q 매트릭스의 크기를 지정하고 정규분포를 가진 랜덤한 값으로 입력합니다. 
    np.random.seed(1)
    P = np.random.normal(scale=1./K, size=(num_reviewers, K))
    Q = np.random.normal(scale=1./K, size=(num_beers, K))

    break_count = 0
       
    # R > 0 인 행 위치, 열 위치, 값을 non_zeros 리스트 객체에 저장. 
    non_zeros = [ (i, j, R[i,j]) for i in range(num_reviewers) for j in range(num_beers) if R[i,j] > 0 ]
   
    # P와 Q 매트릭스를 계속 업데이트(확률적 경사하강법)
    for step in tqdm_notebook(range(steps)):
        errors=[]
        
        for i, j, r in non_zeros:
            # 실제 값과 예측 값의 차이인 오류 값 구함
            eij = r - np.dot(P[i, :], Q[j, :].T)
            # Regularization을 반영한 SGD 업데이트 공식 적용
            P[i,:] = P[i,:] + learning_rate*(eij * Q[j, :] - r_lambda*P[i,:])
            Q[j,:] = Q[j,:] + learning_rate*(eij * P[i, :] - r_lambda*Q[j,:])
            # 정규화 반영하지 않은 식
#             P[i,:] = P[i,:] + 2*learning_rate*(eij * Q[j, :])
#             Q[j,:] = Q[j,:] + 2*learning_rate*(eij * P[i, :])
            
            prediction=np.dot(P[i,:], Q[j,:].T)
#             print(0, step, i, j, prediction)

            errors.append(R[i,j]-prediction)
     
        errors=np.array(errors)
#         print(1, step, errors)
        
        rmse=np.sqrt(np.mean(errors**2))
#         rmse = get_rmse(R, P, Q, non_zeros)
        if (step % 10) == 0 or step==(steps-1):
            print("### iteration step : ", step," rmse : ", rmse)
            
    return P, Q
```

```python
#%%time  
# 실행시간을 마지막에 보여준다
# 경사하강법을 이용한 행렬 분해(4~5분 정도 걸림)
P, Q = matrix_factorization(rating_matrix_new.values, K=10, steps=200, learning_rate=0.01, r_lambda = 0.02)

pred_matrix = np.dot(P, Q.T)
```

```python
ratings_pred_matrix = pd.DataFrame(data=pred_matrix, index= rating_matrix_new.index,
                                   columns = rating_matrix_new.columns)
# 예측 평점 행렬 확인
print(ratings_pred_matrix.shape)
ratings_pred_matrix.head(15)
```

```python
def get_undrunken_beers(rating_matrix_new, userId):
    # 아이디로 입력받은 사용자의 모든 맥주정보 추출하여 Series로 반환함. 
    # 반환된 user_rating 은 맥주명을 index로 가지는 Series 객체임. 
    user_rating = rating_matrix_new.loc[userId,:]
    
    # user_rating이 0보다 크면 기존에 관람한 영화임. 대상 index를 추출하여 list 객체로 만듬
    already_drunken= user_rating[ user_rating > 0].index.tolist()
    
    # 모든 맥주명을 list 객체로 만듬. 
    beers_list = rating_matrix_new.columns.tolist()
    
    # list comprehension으로 already_drunken에 해당하는 맥주는 beers_list에서 제외함. 
    undrunken_list = [ beer for beer in beers_list if beer not in already_drunken]
    
    return undrunken_list
```

```python
def recomm_beers_by_userid(pred_df, userId, undrunken_list, top_n=10):
    # 예측 평점 DataFrame에서 사용자id index와 undrunken_list로 들어온 맥주명 컬럼을 추출하여
    # 가장 예측 평점이 높은 순으로 정렬함. 
    recomm_beers = pred_df.loc[userId, undrunken_list].sort_values(ascending=False)[:top_n]
    return recomm_beers
```

```python
# 사용자가 관람하지 않는 영화명 추출   
undrunken_list = get_undrunken_beers(rating_matrix_new, '007lund(92)')

# 잠재 요인 협업 필터링으로 영화 추천 
recomm_beers = recomm_beers_by_userid(ratings_pred_matrix, '007lund(92)', undrunken_list, top_n=10)

# 평점 데이타를 DataFrame으로 생성. 
recomm_beers = pd.DataFrame(data=recomm_beers.values,index=recomm_beers.index,columns=['pred_score'])
recomm_beers
```
