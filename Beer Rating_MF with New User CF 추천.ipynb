```python
# import numpy as np
# from sklearn.metrics import mean_squared_error
# from tqdm import tqdm_notebook

# def get_rmse(R, P, Q, non_nan):
#     error = 0
#     # 두개의 분해된 행렬 P와 Q.T의 내적 곱으로 예측 R 행렬 생성
#     full_pred_matrix = np.dot(P, Q.T)
    
#     # 실제 R 행렬에서 널이 아닌 값의 위치 인덱스 추출하여 실제 R 행렬과 예측 행렬의 RMSE 추출
#     x_non_nan_ind = [non_nan[0] for non_nan in non_nans]
#     y_non_nan_ind = [non_nan[1] for non_nan in non_nans]
#     R_non_nans = R[x_non_nan_ind, y_non_nan_ind]

#     full_pred_matrix_non_nans = full_pred_matrix[x_non_nan_ind, y_non_nan_ind]
    
#     mse = mean_squared_error(R_non_nans, full_pred_matrix_nan_zeros)
#     rmse = np.sqrt(mse)
    
    
#     return rmse
```

```python
from tqdm import tqdm_notebook

def matrix_factorization(R, K, steps, learning_rate, r_lambda):

    min_rmse=[]
    break_count = 0
  
    num_reviewers, num_beers = R.shape
    # P와 Q 매트릭스의 크기를 지정하고 정규분포를 가진 랜덤한 값으로 입력합니다. 
    np.random.seed(1)
    P = np.random.normal(scale=1./K, size=(num_reviewers, K))
    Q = np.random.normal(scale=1./K, size=(num_beers, K))
      
    # R > 0 인 행 위치, 열 위치, 값을 non_nan 리스트 객체에 저장. 사전에 nan을 0으로 바꿔주던지 nan이 아닌 것만 고르는 것으로 변경이 바람직
    non_nan = [ (i, j, R[i,j]) for i in range(num_reviewers) for j in range(num_beers) if R[i,j] > 0 ]
    
#     #사용자와 맥주의 경향성을 제외하기 위한 bias 값 0으로 초기화 ------------
#     #사용자 평가경향이기 때문에 크기는 리뷰어 수와 동일
#     user_bias=np.zeros(num_reviewers)
#     #아이템 평가경향이기 때문에 크기는 맥주 수와 동일
#     item_bias=np.zeros(num_beers)
#     #전체 평점 평균
#     total_mean=np.mean(R[R > 0]) #---> non zero가 아닌 nan을 뽑아 평균을 구해야 함...
#     #-------------------------------------------------------------------------
    
    # P와 Q 매트릭스를 계속 업데이트(확률적 경사하강법)
    for step in tqdm_notebook(range(steps)):
        errors=[]
        
        for i, j, r in non_nan:
            # 실제 값과 예측 값의 차이인 오류 값 구함
            eij = r - np.dot(P[i, :], Q[j, :].T)
            # Regularization을 반영한 SGD 업데이트 공식 적용: 오차를 줄여주는 새로운 P, Q를 구하기 위한 식
            P[i,:] = P[i,:] + learning_rate*(eij * Q[j, :] - r_lambda*P[i,:])
            Q[j,:] = Q[j,:] + learning_rate*(eij * P[i, :] - r_lambda*Q[j,:])

            # 정규화 반영하지 않은 식: 오차를 줄여주는 새로운 P, Q를 구하기 위한 식
#             P[i,:] = P[i,:] + 2*learning_rate*(eij * Q[j, :])
#             Q[j,:] = Q[j,:] + 2*learning_rate*(eij * P[i, :])


#             #사용자와 맥주의 경향성을 제외하기 위한 공식 ----------------------
#             user_bias[i]=user_bias[i]+learning_rate*(eij-r_lambda*user_bias[i])
#             item_bias[j]=item_bias[j]+learning_rate*(eij-r_lambda*item_bias[j])
            
#             #사용자와 맥주의 경향성을 제외한 예측치 산출식
#             prediction=total_mean+user_bias[i]+item_bias[j]+np.dot(P[i,:], Q[j,:].T)
#             #-------------------------------------------------------------------
            
            prediction=np.dot(P[i,:], Q[j,:].T)
            
            # 실제 값과 새로운 예측 값의 차이인 오류 값 구함
            errors.append(R[i,j]-prediction)
     
        errors=np.array(errors)
        
        #RMSE 값 산출
        rmse=np.sqrt(np.mean(errors**2))

#        rmse = get_rmse(R, P, Q, non_zeros)
        
        #iteration 이 50번씩 돌때와 마지막에 iteration과 RMSE 출력
        if (step % 50) == 0 or step==(steps-1):
            print("### iteration step : ", step," rmse : ", rmse)
        
#         # 최적 K값 찾을 때 열기---------
#         k_result.append(k)
#         step_result.append(step)
#         rmse_result.append(rmse)
          #--------------------------------
    
#     # 최적 K값 + iteration(Step)값 찾을 때 열기 : 각 K별, iteration별 RMSE가 최저값을 나타내는 K, iteration과 RMSE를 summary에 저장---   
#     min_rmse=np.min(rmse)
#     min_index=rmse_result.index(min_rmse)
#     summary.append([k_result[min_index], step_result[min_index], rmse_result[min_index]])
      #----------------------------------------------------------------------------------------------------------------------------------

        
    return P, Q, rmse, step
```

```python
#%%time  
# 실행시간을 마지막에 보여준다
# 경사하강법을 이용한 행렬 분해(10분 정도 걸림)
    
P, Q, rmse, step = matrix_factorization(rating_matrix_new.values, K=100, steps=500, learning_rate=0.01, r_lambda = 0.01)

pred_matrix = np.dot(P, Q.T)
```

```python
#새로 입력한 유저의 맥주 평점을 기반으로 먹어보지 않은 리스트 추출

def get_undrunken_beers(rating_matrix_new_user, userId):
    # 아이디로 입력받은 사용자의 모든 맥주정보 추출하여 Series로 반환함. 
    # 반환된 user_rating 은 맥주명을 index로 가지는 Series 객체임. 
    user_rating = rating_matrix_new_user.loc[userId,:]
    
    # user_rating이 0보다 크면 기존에 마셔본 맥주임. 대상 index를 추출하여 list 객체로 만듬
    already_drunken= user_rating[ user_rating > 0].index.tolist()
    
    # 모든 맥주명을 list 객체로 만듬. 
    beers_list = rating_matrix_new_user.columns.tolist()
    
    # list comprehension으로 already_drunken에 해당하는 맥주는 beers_list에서 제외함. 
    undrunken_list = [ beer for beer in beers_list if beer not in already_drunken]
    
    return undrunken_list
```

```python
#새 유저와 가장 유사한 유저의 고평점 예측 맥주 10개 추출(새 유저의 평점을 준 맥주는 제외)

def recomm_beers_by_userid(pred_df, userId, undrunken_list, top_n=10):
    # 예측 평점 DataFrame에서 사용자id index와 undrunken_list로 들어온 맥주명 컬럼을 추출하여
    # 가장 예측 평점이 높은 순으로 정렬함. 
    recomm_beers = pred_df.loc[userId, undrunken_list].sort_values(ascending=False)[:top_n]
    return recomm_beers
```

```python
# 사용자가 마셔보지지 않는 맥주명 추출   
undrunken_list = get_undrunken_beers(rating_matrix_new_user, user_id)
len(undrunken_list)

# 잠재 요인 협업 필터링으로 맥주 10개 추천 
recomm_beers = recomm_beers_by_userid(ratings_pred_matrix, user_sim_df_sorted.index[1], undrunken_list, top_n=10)

# 평점 데이타를 DataFrame으로 생성. 
recomm_beers = pd.DataFrame(data=recomm_beers.values,index=recomm_beers.index,columns=['pred_score'])
recomm_beers
```
