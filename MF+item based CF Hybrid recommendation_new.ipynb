# Matrix Factorization

```python
from tqdm import tqdm_notebook

def matrix_factorization(R, K, steps, learning_rate, r_lambda):

    min_rmse=[]
    break_count = 0
  
    num_reviewers, num_beers = R.shape
    # P와 Q 매트릭스의 크기를 지정하고 정규분포를 가진 랜덤한 값으로 입력합니다. 
    np.random.seed(1)
    P = np.random.normal(scale=1./K, size=(num_reviewers, K))
    Q = np.random.normal(scale=1./K, size=(num_beers, K))
      
    # R > 0 인 행 위치, 열 위치, 값을 non_nan 리스트 객체에 저장. 사전에 nan을 0으로 바꿔주던지 nan이 아닌 것만 고르는 것으로 변경이 바람직
    non_nan = [ (i, j, R[i,j]) for i in range(num_reviewers) for j in range(num_beers) if R[i,j] > 0 ]
   
   
    # P와 Q 매트릭스를 계속 업데이트(확률적 경사하강법)
    for step in tqdm_notebook(range(steps)):
        errors=[]
        
        for i, j, r in non_nan:
            # 실제 값과 예측 값의 차이인 오류 값 구함
            eij = r - np.dot(P[i, :], Q[j, :].T)
            # Regularization을 반영한 SGD 업데이트 공식 적용: 오차를 줄여주는 새로운 P, Q를 구하기 위한 식
            P[i,:] = P[i,:] + learning_rate*(eij * Q[j, :] - r_lambda*P[i,:])
            Q[j,:] = Q[j,:] + learning_rate*(eij * P[i, :] - r_lambda*Q[j,:])
            prediction=np.dot(P[i,:], Q[j,:].T)
            
            # 실제 값과 새로운 예측 값의 차이인 오류 값 구함
            errors.append(R[i,j]-prediction)
     
        errors=np.array(errors)
        
        #RMSE 값 산출
        rmse=np.sqrt(np.mean(errors**2))

        #iteration 이 50번씩 돌때와 마지막에 iteration과 RMSE 출력
        if (step % 50) == 0 or step==(steps-1):
            print("### iteration step : ", step," rmse : ", rmse)
            
    return P, Q, rmse, step
```

```python
#%%time  
# 실행시간을 마지막에 보여준다
# 경사하강법을 이용한 행렬 분해(10분 정도 걸림)
    
P, Q, rmse, step = matrix_factorization(rating_matrix_new.values, K=100, steps=500, learning_rate=0.01, r_lambda = 0.01)

pred_matrix = np.dot(P, Q.T)
```

# Item-based CF

```python
# 아이템 기반 유사도 측정

from sklearn.metrics.pairwise import cosine_similarity

rating_matrix_new_T=np.transpose(rating_matrix_new)
matrix_dummy=rating_matrix_new_T.copy().fillna(0)

item_similarity=cosine_similarity(matrix_dummy, matrix_dummy)
item_similarity=pd.DataFrame(item_similarity, index=rating_matrix_new_T.index, columns=rating_matrix_new_T.index)
```

```python
item_similarity
```

```python
# 아이템기반 유사도와 실제값과의 내적을 통해 구한 예측 평점을 각각 구하고 RMSE 값 측정 

from tqdm import tqdm_notebook

#사용자와 맥주를 pair로 짝 맞춤
id_pairs=zip(df_new['아이디'],df_new['맥주'], df_new['평점'])
pred_ratings=[]
real_ratings=[]

for (user, beer, rating) in tqdm_notebook(id_pairs):
#     print(user, beer)
    # 현재 맥주와 다른 맥주의 similarity 값 가져오기
    sim_scores = item_similarity[beer].copy()
    # 현 사용자의 모든 rating 값 가져오기
    user_ratings= rating_matrix_new.loc[user].copy()
    # 사용자가 평가하지 않은 맥주 index 가져오기
    none_rating_idx= user_ratings[user_ratings.isnull()].index
    # 사용자가 평가하지 않은 맥주 제거
    user_ratings = user_ratings.dropna()
    # 사용자가 평가하지 않은 맥주의 similarity 값 제거
    sim_scores = sim_scores.drop(none_rating_idx)
    # 현 맥주에 대한 예상 rating 계산, 가중치는 현 맥주와 사용자가 평가한 맥주의 유사도
#     print(user_ratings)
#     print(sim_scores)
    pred_rating =np.dot(sim_scores, user_ratings) / sim_scores.sum()
#     print(pred_rating)
    pred_ratings.append(pred_rating)
#     print(pred_ratings)
    # 현 사용자의 현재 맥주의 실제 평점
    real_ratings.append(rating)
    
rmse=np.sqrt(np.mean((np.array(real_ratings)-np.array(pred_ratings))**2))
print(rmse)
```
