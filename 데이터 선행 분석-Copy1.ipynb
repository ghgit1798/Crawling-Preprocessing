```python
#연도, 월별 그룹지어서 평균을 내고 연도, 월 붙임

df_yearmonthly_avg=df.groupby(['연도', '월'], as_index=False)['평점'].mean()
df_yearmonthly_avg['연월']=df_yearmonthly_avg['연도']+df_yearmonthly_avg['월']
df_yearmonthly_avg
```

```python
#연월별 평점 평균

df_yearmonthly_avg.plot(x='연월', y='평점')
plt.show()

plt.scatter(data=df_yearmonthly_avg, x='연월', y='평점')
plt.xlabel('연월')
plt.ylabel('평점')
plt.title('평점 추세')
plt.show()

# import seaborn as sns
# sns.regplot(x='연월',y='평점',data=df_yearmonthly_avg, fit_reg=True)
# plt.show()
```
